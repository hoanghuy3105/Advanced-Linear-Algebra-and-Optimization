```python
from scipy import optimize
import numpy as np
```

```python
#Scipy Optimazation Functions include following methods such as 
#Nelder-Mead, Powell, CG, BFGS, Newton-CG, TNC, SLSQP
```

```python
#I would create a function that includes all of these functions
#so that i could easily test them at the same time with different points
```

```python
L = ['Nelder-Mead', 'Powell', 'CG', 'BFGS', 'TNC', 'SLSQP']
```

```python
def Opt_Function(List, function, points):
    for i in List:
        optimization = optimize.minimize(function, points, method = i)
        print('\n')
        print(i+ " Method")
        print("------------------")
        print('\n')
        print(optimization)
        print('Iterations : ', optimization['nit'])
```

```python
def F_Rosen(x): 
    return 100.0*(x[1] - x[0]**2)**2 + (1 - x[0])**2
```

```python
# I would start to test Rosen's Function from [-1,1]
```

```python
Opt_Function(L, F_Rosen, [-1,1])
```

```python
# I would start to test Rosen's Function from [-100,100]
```

```python
Opt_Function(L, F_Rosen, [-100,100])
```

```python
# I would start to test Rosen's Function from [-1000,1000]
```

```python
Opt_Function(L, F_Rosen, [-1000,1000])
```

```python
def F_Booth(x):
    return ((x[0]+2*x[1]-7)**2) + (2*x[0]+x[1]-5)**2
```

```python
# I would start to test Booth's Function from [-1,1]
```

```python
Opt_Function(L, F_Booth, [-1,1])
```
